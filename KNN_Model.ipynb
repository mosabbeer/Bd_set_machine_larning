```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
url ='/content/drive/MyDrive/Dataset_1_KNN.xlsx'
df=pd.read_excel('/content/drive/MyDrive/Dataset_1_KNN.xlsx')
df
```

```python
df.head()
df.tail()
```

```python
df['Draft']=df['Draft'].map({'yes':1, 'no':0})
df.head()
```

```python
#feature and lavel
x=df[['Speed','Agility']].values
y=df['Draft'].values
```

```python
k=3
knn=KNeighborsClassifier(n_neighbors=k)
knn.fit(x,y)
```

```python
speed=float(input('Enter the speed:'))
agility=float(input('Enter the agility:'))
new_student=np.array([[speed,agility]])
prediction=knn.predict(new_student)
print('Draft:',prediction)
```

```python
print(f"Predicted class for the sudent with speed ={speed} and agility ={agility} : {'seleclected' if prediction[0]==1 else 'we selected'}")
```
